## Filter 4 of the largest cities into a new dataframe

```python
# copy old df and create new dataset
df2 = df.copy()
```

```python
# make filter to get cities
largest_cities = (df2['SizeRank'] >= 0) & (df2['SizeRank'] <=3)
```

```python
# print data frame
df2[largest_cities]
```

```python
# make filter
df2 = df2[largest_cities]
df2
```

## Melt the data to long-form and prepare it for time series analysis

```python
# display columns
df2.columns
```

```python
# melt data
melt = pd.melt(df2.drop(columns=['RegionID','StateName']),
                 id_vars=['SizeRank', 'RegionName','RegionType','State','Metro','CountyName'],
                value_name='Home Value',var_name="Date")
melt
```

## Convert melted dates into a datetime datatype

```python
# convert melt to datetime
melt['Date'] = pd.to_datetime(melt['Date'])
melt.info()
```

```python
# Make datetime column the index
melt = melt.set_index("Date")
melt
```

## Resample dataframe as monthly frequency, group by City

```python
df2_rs = melt.groupby('RegionName').resample("M").sum()
df2_rs
```

```python
# Func to convert thousands into K
def thousands(x,pos):
    """function for use wth matplotlib FuncFormatter -  formats money in millions"""
    new_x = x/1_000
    return f"{new_x:,.0f}K"

# Create the formatter
fmt_thous = mticks.FuncFormatter(thousands)
```
